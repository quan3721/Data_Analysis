```python
# ansewring questions about branch
# Which location has highest and the lowest sales ?
# represent the sales on a bar chart, also show the market share for each location using a pie chart
```

```python
# -- Create a variable to contain data from function groupby 'Location' column -- #

location_list = sales.groupby('Location')
```

```python
# -- Show the information of each unique values from Location column -- #

for x,y in location_list:
    print(y)
```

```python
# -- Show the unique value from Location column -- #

for x,y in location_list:
    print(x)
```

```python
# -- Create a list to contain the unique value from variable contained data -- #
# -- after executed Function groupby -- #

location = [x for x,y in location_list]
location
```

```python
# -- Get the Sum value of Total column following each unique values from Location column  -- #

sales.groupby('Location').sum()['Total']
```

```python
# -- Import Matplotlib Library -- #

import matplotlib.pyplot as plt
```

```python
# -- Draw a bar chart with x dimension : Unique value of Location column and y dimension : The Sum of
# -- value of Total column following each Unique value of Location column

plt.bar(location, sales.groupby('Location').sum()['Total'])
plt.show()
```

```python
# -- Draw a PIE Chart show The Sum of value of Total column 
# -- following each Unique value of Location column

plt.pie(sales.groupby('Location').sum()['Total'], labels=location, autopct='%1.1f%%') 
# note : autopct='%1.1f%%' : show the percentage
plt.show()
```
